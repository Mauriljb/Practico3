```python
el=np.array(gal.elliptical)
es=np.array(gal.spiral)
nel, nes=0,0
for i in range(len(el)):
    if el[i]==1:
        nel=nel+1
    else:
        nes=nes+1
print(nel,nes)
```

```python
chi=((nel-46094)**2)/46094+((nes-46094)**2/46094)
chit=2.71
if chi<chit:
    print('No se puede rechazar la hipÃ³tesis nula, nuestra muestra es consistente con una distribucoin uniforme')
else:
    print('La hipotesis nula fue descartada, la muestra NO es consistente con una distribucion uniforme')
```
